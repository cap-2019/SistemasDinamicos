1. First ﬁx a =1. Use the graph of f1,b to construct the bifurcation diagram for this family of differential equations depending on b. 

```python
def solEq(a,beta):
    from sympy import symbols, solve
    x = symbols('x')
    expr = a*x-x**3-beta
    sol = solve(expr)
    return sol

def model(y,t,a,b):
    dydt = (a*y-y**3) - b
    return dydt

from pylab import *
xmax = 3.0
xmin = -xmax
D = 19
ymax = 3.0
ymin = -ymax

x = linspace(ymin, ymax, D)#B
y = linspace(ymin, ymax, D)
X, Y = meshgrid(x, y)
        
def plotEq(a):
    deg = arctan(a*Y-Y**3 - X)
    QP = quiver(X,Y,cos(deg),sin(deg))
    t = np.linspace(-2, 2)
    plt.plot(model(t,t,a,0),t, 'g',lw=1,label='a='+str(a))
    plt.xlabel('B')
    plt.ylabel('X')
    plt.title('A='+str(a))
    
    solu=solEq(a,beta=0)
    print(solu)
    for s in solu:
        try:
            plot(0,s,'bo')
        except:
            None
    plt.xlim(-2,2)

        

        
a=1
plotEq(a)

plt.show()
```

4. Now ﬁx b and use the graph to construct the bifurcation diagram for this family, which this time depends on a. 

```python
from pylab import *
xmax = 3.0
xmin = -xmax
D = 19
ymax = 3.0
ymin = -ymax

x = linspace(ymin, ymax, D)#B
y = linspace(ymin, ymax, D)
X, Y = meshgrid(x, y)
a=0
def solEq(a,beta):
    from sympy import symbols, solve
    x = symbols('x')
    expr = a*x-x**3-beta
    sol = solve(expr)
    return sol

def model(y,t,a,b):
    dydt = (a*y-y**3) - b
    return dydt

def plotEqB(b):
    deg = arctan(X*Y-Y**3 - b)
    QP = quiver(X,Y,cos(deg),sin(deg))
    t = np.linspace(-2, 2)
    plt.plot(model(t,t,a,b),t, 'g',lw=1,label='a='+str(a)+" b="+str(b))
    plt.xlabel('A')
    plt.ylabel('X')
    plt.title('B='+str(b))
    
    solu=solEq(a=0,beta=b)
    for s in solu:
        try:
            plot(0,s,'bo')
        except:
            None
    plt.xlim(-2,2)
    plt.show()
```

```python
def solEq(t):
    from sympy import symbols, solve
    x = symbols('x')
    expr= x+sym.cos(t)
    sol = solve(expr)
    return sol

def model(y,t):
    dydt = y+cos(t)
    return dydt

def plotModel(ic):
    tn = np.linspace(0, 2*pi, 21)
    y1 = odeint(model, ic, tn)
    plt.plot(tn, y1, color='b')

ic=[.5,0,-.5]
for i in ic:
    plotModel(i)

times = np.linspace(0, 7,  21)
x = np.linspace(-10,10, 100)
T, X = np.meshgrid(times, x)   # make a grid that roughly matches the Strogatz grid
dxdt = X+cos(T)            # the equation of interest
dt = .5*np.ones(X.shape)  # a constant value (.5 is just so segments don't run into each other -- given spacing of times array
dx = dxdt * dt            # given dt, now calc dx for the line segment

plt.quiver(T, X, dt, dx, headwidth=3., angles='xy', scale=15.)
for t in times:
    solu=solEq(t)
    for s in solu:
        try:
            plot(t,s,'go')
        except:
            None
plt.ylim(top=2)  
plt.ylim(bottom=-2)  
plt.xlim(left=0)  
plt.xlim(right=7)  
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```
